```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# Import XGBClassifier
from xgboost import XGBClassifier
# Import the classification metrics
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
from sklearn.metrics import plot_confusion_matrix
# Use GridSearchCV to tune the model's parameters
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostClassifier
import joblib
```

```
data_df = pd.read_csv('/content/drive/My Drive/heart_failure_dataset.csv')
data_df
```

```
data_df.info()
```

```
# Create the X variable
X = data_df.drop(["DEATH_EVENT"],axis=1)
```

```
# Create the y variable
y = data_df["DEATH_EVENT"]
```

```
# Now split the data into a training and testing set
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

```
scaler = MinMaxScaler()
scaler
```

```
# Scale the data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```
# Create an instance of XGBClassifier
model = XGBClassifier()
# Fit the data to the training set
model.fit(X_train_scaled,y_train) 
```

```
# Now lets make predctions
predictions = model.predict(X_test_scaled)
```

```
# Check the model's accuracy
accuracy_score(predictions, y_test)
```

```
# Create an instance of XGBClassifier
CatBoost_model = CatBoostClassifier()
# Fit the data to the training set
CatBoost_model.fit(X_train_scaled,y_train) 
```

```
# Now lets make predctions
CATpredictions = CatBoost_model.predict(X_test_scaled)
# Check the model's accuracy
accuracy_score(CATpredictions, y_test)
```
